# baidu spider

```python
def baidu():
    url = 'https://top.baidu.com/board?tab=realtime'

    baidu_hot = []
    soup = BeautifulSoup(requests.get(url).text, 'html.parser')
    for i in soup.find(class_='rel container_2VTvm').find(class_='container right-container_2EFJr').findAll(
            class_='category-wrap_iQLoo horizontal_1eKyQ'):
        tittle = i.find(class_="c-single-text-ellipsis").text
        baidu_hot.append(tittle.strip())

    return baidu_hot
```

```python
def zhihu_q():
    url = 'https://www.zhihu.com/api/v3/feed/topstory/hot-lists/total?limit=100'

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.125 Safari/537.36'
    }
    response = requests.get(url, headers=headers)

    zhihu_q_hot = {}

    for i in json.loads(response.text)['data']:
        zhihu_q_hot[i['target']['title'].strip()] = {'href': f"https://www.zhihu.com/question/{i['target']['id']}",
                                                     'hot': i['detail_text']}
    return zhihu_q_hot

```

# toutiao spider

```python
def toutiao():
    url = 'https://is-lq.snssdk.com/api/suggest_words/?business_id=10016'
    response = requests.get(url)
    toutiao_hot = []
    for i in json.loads(response.text)['data'][0]['words']:
        toutiao_hot.append(i['word'])
    return toutiao_hot


```

```python


def douyin():
    url = 'https://aweme.snssdk.com/aweme/v1/hot/search/list/'

    HEADERS = {
        'user-agent': 'okhttp3'
    }
    response = requests.get(url, headers=HEADERS)
    douyin_hot = []

    for i in json.loads(response.text)['data']['word_list']:
        douyin_hot.append(i['word'])
    return douyin_hot

```
